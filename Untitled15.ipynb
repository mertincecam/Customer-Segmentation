```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np

#(Sahte Veri)
np.random.seed(42)
data = {
    'Musteri_ID': range(1, 201),
    'Yillik_Gelir_binTL': np.random.randint(20, 120, 200),
    'Harcama_Skoru_1_100': np.random.randint(1, 100, 200)
}
df = pd.DataFrame(data)

X = df[['Yillik_Gelir_binTL', 'Harcama_Skoru_1_100']]
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)
df['Segment'] = kmeans.fit_predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Yillik_Gelir_binTL', y='Harcama_Skoru_1_100',
                hue='Segment', palette='viridis', s=100)
plt.title('Müşteri Segmentasyonu Analizi')
plt.xlabel('Yıllık Gelir (bin TL)')
plt.ylabel('Harcama Skoru (1-100)')
plt.legend(title='Segmentler')
plt.show()

df.to_csv('musteri_segmentasyonu.csv', index=False)
```
